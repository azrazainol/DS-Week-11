# **A quick glance on Matplotlib**
- a 2D plotting library for Python that produces high-quality static plots

```python
# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
plt.plot(x, y)

# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("simple plot"); #use semicolon (;) at the end to suppress the output message
```

# **A quick glance on Seaborn**
- built on top of Matplotlib
- provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a DataFrame using the data
data = {"x" : x, "y" : y}
df = pd.DataFrame(data)
df
# Create a scatter plot using Seaborn
sns.scatterplot(x = "x", y = "y", data = df)

# Add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Seaborn Scatterplot");
```

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3.   **dpi stands for dot per inch

### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# change the value of wspace and hspace
# sharex=True parameter means that the subplots will share the same x-axis
# The "fig" variable is a reference to the entire figure
# The "axes" is a 2D array of subplot references
fig, axes = plt.subplots(2, 2, sharex = True, sharey = False)

# The outer loop (for i in range(2)) iterates over rows
# The inner loop (for j in range(2)) iterates over columns
# Top left -> Top right -> bottom left -> bottom right
# (0,0) -> top left
# (0,1) -> top right
# (1,0) -> bottom left
# (1,1) -> bottm right
for i in range(2):
  for j in range(2):
    axes[i, j].hist(rng.standard_normal(500), bins = 50, color = "navy", alpha = 0.5)

# Adjust the space around the subplot
# wspace = width of the subplots
# hspace = height of the padding spaces
fig.subplots_adjust(wspace = 0.5, hspace = 0.5)
```

# **Adding shapes to a plot**
- **Important for final exam**


```python
# Adding shapes to a plot
fig, ax = plt.subplots() #always use this code for creating subplots

#create a rectangle shape
# .4 -> width of the x-axis
# .2 -> 7.5 + .2 = .9 -> height of the rectangle
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.2, color = "#483d8b", alpha = 0.3)

# create a circle shape
# .15 -> radius/width of the circle
circ = plt.Circle((0.7, 0.2), 0.15, color = "#556b2f", alpha = 0.3)

# create a triangle shape
# hexagon  -> polygon of 6 sides
# so the more points you add in the points part, the more angles
# as for this one, you put 3 so it becomes 3 sides = triangle
trig = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = "#9370db", alpha = 0.5)

# add all together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(trig);
```

```python
# Adding shapes to a plot : Attention
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color="green", alpha=0.5)

# create a square shape
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color="black", alpha=0.9)


# create an ellipse shpae
# (0.5, 0.5) -> This tuple specifies the center of the ellipse
# 0.2 -> width of the ellipse
# 0.1 -> height of the ellipse
ellipse = mpatches.Ellipse((0.5, 0.5), 0.2, 0.1, color = "pink", alpha = 0.7)

# add hexagon
# 6 -> specify number of sides in the regular polygon
# 0.1 -> distance from the center of the circumcircle (the point at the center of the hexagon) to any of its vertices
hex = mpatches.RegularPolygon((0.7, 0.8), 6, 0.1, color = "red")

# add all the subplots together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```

```python
# A little project: dataset about restaurant tipping
# Download the tips data herehttps://bit.ly/3jCROII:
tips = pd.read_csv("https://bit.ly/3jCROII")
tips
```

```python
# Using the crosstab() function
# To compute a simple cross-tabulation of two (or more) factors.
# Create a two factors data frame
party_counts = pd.crosstab(tips["day"], tips["size"])
party_counts
```

```python
# Extract party_count from col 2 to 5
party_counts_edit = party_counts.loc[:, 2:5]
party_counts_edit
```

```python
# Normalize each row to sum to 1
# axis="index" argument indicates that the division should be done along the rows
# Doing some data wrangling for data frame
party_pcts = party_counts_edit.div(party_counts_edit.sum(axis = "columns"), axis = "index")
party_pcts
```
